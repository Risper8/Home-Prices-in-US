## Modelling

#### Importing required libraries

```python

# Feature Engineering
from sklearn.preprocessing import StandardScaler, LabelEncoder #feature scaling

# Machine Learning Models
from statsmodels.stats.outliers_influence import variance_inflation_factor #handling outliers
from sklearn.linear_model import LinearRegression #linear model 
from sklearn.tree import DecisionTreeRegressor #Tree model for regression
from sklearn.ensemble import RandomForestRegressor #Tree boasting model for regression

# Model Selection and Validation
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#split data into training and test data

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

```

### Standardizing the data(Feature Scaling)

Standardized the original data to follow a normal distribution,i.e, have a mean 0 and a variance of 1

```python
#standard scaler
sc = StandardScaler()

#fit the scaler to the data (compute mean and standard deviation)
sc.fit(x)

# transfor training and test data to standardized them
x_train_standardized = sc.transform(x_train)
x_test_standardized = sc.transform(x_test)
```

#### Linear Regression

```python

import matplotlib.pyplot as plt

# Train the regression model
regressor = LinearRegression()
regressor.fit(x_train_standardized, y_train)

# Make predictions on the test set
y_pred = regressor.predict(x_test_standardized)

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the results
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)

```

#### Random Forest Regressor

```python

# Instantiate the Random Forest Regressor model 
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_regressor.fit(x_train_standardized, y_train)

# Make predictions
y_pred = rf_regressor.predict(x_test_standardized)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

#### Decision Tree Regression

```python
dt_regressor = DecisionTreeRegressor()

#Fit the model
dt_regressor.fit(x_train_standardized, y_train)

# Make predictions
y_pred = dt_regressor.predict(x_test_standardized)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

From these results, the Random Forest model has the lowest MAE and MSE values, indicating better performance in terms of prediction accuracy. Additionally, it has the highest R-squared value, suggesting a better fit to the data compared to the other models.

Therefore, based on the provided evaluation metrics, the Random Forest model appears to be the best choice for predicting your data.
